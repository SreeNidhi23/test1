```python
l=["r","b","y"]
k=[".","v","1","o"]
d=["-","--","-",",",":"]
x=range(2,22,2)
y=range(3,33,3)
x1=np.array(x)*2
y1=np.array(y)*3
x2=np.array(x)**2
y2=np.array(y)**2
for i in l:
    for j in k:
        for p in range(10,15):
            for n in d:
                    plt.plot(x,y,color=i,marker=j,markersize=p,ls=n)
                    plt.plot(x1,y1,color=i,marker=j,markersize=p,ls=n)
                    plt.plot(x2,y2,color=i,marker=j,markersize=p,ls=n)
                    plt.ylabel("Numbers")
                    plt.xlabel('Indices')
                    plt.title('MyPlot')
                    plt.grid()
                    plt.show()
```

```python
import numpy as np
t= np.arange(0.,5.,0.2)
for i in range(1,6):
    plt.plot(t,t**2, 'b--', label='^2')
    plt.plot(t,t**2.2, 'rs', label='^2.2')
    plt.plot(t,t**2.5, 'g^', label='^2.5')
    plt.grid()
    plt.xlabel("x axis")
    plt.ylabel("y axis")
    plt.legend(loc=i)
    plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0,5.0,0.1)
t2= np.arange(0.0,5.0,0.02)
plt.figure(1,figsize=(8,10))
plt.subplot(2,2,1)
plt.grid()
plt.plot(t1, f(t1), 'y-')
plt.subplot(2,2,2)
plt.grid()
plt.plot(t1, f(t1), 'm-')
plt.subplot(2,2,4)
plt.grid()
plt.plot(t1, f(t1), 'g-')
plt.subplot(2,2,3)
plt.plot(t2, np.cos(2*np.pi*t2),'r--')
plt.show()
```
