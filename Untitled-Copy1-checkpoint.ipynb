```python
df.columns=(['State_UT', 'Total_Cases', 'Active', 'Discharged', 'Deaths','Active_Ratio', 'Discharge_Ratio', 'Death_Ratio','Population'])
```

```python
def num2MB(num):
    '''
        num: int, float
        it will return values like thousands(10K), Millions(10M),Billions(1B)
        '''
    if num < 1000: return int(num)
    if num >= 1000 and num < 1000000:return f'{float("%.1f" %(num/1000))}K'
    elif num>= 1000000 and num < 1000000000: return f'{float("%.1f" %(num/1000000))}M'
    else: return f'{float("%.1f" %(num/1000000000))}B'
    
def bar_Value(data,data2=None,title=None,x_pos=0.25,y_pos=10):
    '''
    data: data
    data2: hue data
    title: str- title of graph
    x_pos: int/float - to adjust x-axis position of text
     y_pos=10: int/float - to adjust y-axis position of text
    
    '''
    ax = sns.barplot(data,data2)
    for i in ax.patches:
        height = i.get_height() # get height of bar (value of y axis)      
        a = num2MB(height)
        x = i.get_x() # get x_axis value
        ax.text(x=x+x_pos,y=height+y_pos,s=a,rotation=None) # shift+tab for more parameter
    plt.title(title,fontdict={'fontsize':20,'color':'red'})
    plt.xticks(rotation=90)
```

```python
temp = df.sort_values(by='Total_Cases',ascending=False).head(10)
temp2 = df.sort_values(by='Total_Cases').head(10)


plt.figure(figsize=(20,6))
plt.subplot(121)
bar_Value(temp['State_UT'],temp['Total_Cases'],x_pos=0,y_pos=50000,title='Top 10 State: Highest Confirmed Cases')

plt.subplot(122)
bar_Value(temp2['State_UT'],temp2['Total_Cases'],x_pos=0,y_pos=500,title='Top 10 State: Lowest Confirmed Cases')
```
